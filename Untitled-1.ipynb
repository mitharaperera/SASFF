```python
#sobel vertical
%matplotlib inline
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

img=cv.imread('contact_lens.tif',cv.IMREAD_REDUCED_GRAYSCALE_2)

kernel=np.array([(-1,-2,-1),(0,0,0),(1,2,1)],np.float32)
imgc=cv.filter2D(img.astype('float'),-1,kernel)
imgc=cv.normalize(imgc.astype('float'),None,0,255.0,cv.NORM_MINMAX)

fig,axes=plt.subplots(1,2,sharex='all',sharey='all',figsize=(18,18))
axes[0].imshow(img,cmap='gray',vmin=0,vmax=255)
axes[0].set_title('Original')
axes[1].imshow(imgc,cmap='gray',vmin=0,vmax=255)
axes[1].set_title('Sobel Vertical')
plt.show()
```

```python
#Sobel Horizontal
%matplotlib inline 
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
img = cv.imread('contact_lens.tif',cv.IMREAD_REDUCED_GRAYSCALE_2)

kernel = np.array([(-1,0,-1),(-2,0,2),(-1,0,1)],dtype='float')
imgc=cv.filter2D(img,-1,kernel)
imgc=cv.normalize(imgc.astype('float'),None,0,255.0,cv.NORM_MINMAX)

fig,axes=plt.subplots(1,2,sharex='all',sharey='all',figsize=(18,18))

axes[0].imshow(img,cmap='gray')
axes[0].set_title('Orginal')
axes[0].set_xticks([]),axes[0].set_yticks([])
axes[1].imshow(imgc,cmap='gray')
axes[1].set_title('Sobel Horizontal')
axes[1].set_xticks([]),axes[1].set_yticks([])
plt.show()
```
