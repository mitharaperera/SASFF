```python
#sobel vertical
%matplotlib inline
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

img=cv.imread('contact_lens.tif',cv.IMREAD_REDUCED_GRAYSCALE_2)
assert img is not None
s_verti=np.array([(-1,-2,-1),(0,0,0),(1,2,1)],np.float32)
s_hori = np.array([(-1,0,1),(-2,0,2),(-1,0,1)],np.float32)

imgv=cv.filter2D(img.astype('float'),-1,s_verti)
imgh=cv.filter2D(img.astype('float'),-1,s_hori)

imgv1=cv.normalize(imgv.astype('float'),None,0,255.0,cv.NORM_MINMAX)
imgh1=cv.normalize(imgh.astype('float'),None,0,255.0,cv.NORM_MINMAX)

fig,axes=plt.subplots(1,3,sharex='all',sharey='all',figsize=(18,18))
axes[0].imshow(img,cmap='gray',vmin=0,vmax=255)
axes[0].set_title('Original')
axes[1].imshow(imgv1,cmap='gray',vmin=0,vmax=255)
axes[1].set_title('Sobel Vertical')
axes[2].imshow(imgh1,cmap='gray',vmin=0,vmax=255)
axes[2].set_title('Sobel Horizontal')
plt.show()
```

```python
#Sobel Horizontal
%matplotlib inline 
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
img = cv.imread('contact_lens.tif',cv.IMREAD_REDUCED_GRAYSCALE_2)

kernel = np.array([(-1,0,-1),(-2,0,2),(-1,0,1)],dtype='float')
imgc=cv.filter2D(img,-1,kernel)
imgc=cv.normalize(imgc.astype('float'),None,0,255.0,cv.NORM_MINMAX)

fig,axes=plt.subplots(1,2,sharex='all',sharey='all',figsize=(18,18))

axes[0].imshow(img,cmap='gray')
axes[0].set_title('Orginal')
axes[0].set_xticks([]),axes[0].set_yticks([])
axes[1].imshow(imgc,cmap='gray')
axes[1].set_title('Sobel Horizontal')
axes[1].set_xticks([]),axes[1].set_yticks([])
plt.show()
```

```python
import cv2
import numpy as np

# Load the image
img = cv2.imread('contact_lens.tif', 0)

# Compute the gradient in the x and y direction
gx = cv2.Sobel(img, cv2.CV_32F, 1, 0)
gy = cv2.Sobel(img, cv2.CV_32F, 0, 1)

# Compute the gradient magnitude and normalize the result
magnitude1 = np.linalg.norm(gx)
magnitude2 = np.linalg.norm(gy)
mag = np.add(magnitude1, magnitude2)
mag = cv2.normalize(mag, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U)

# Display the result
cv2.imshow('Gradient Magnitude', mag)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
